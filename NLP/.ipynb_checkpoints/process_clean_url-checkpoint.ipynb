```python
import numpy as np
import pandas as pd

#1. import the data, find the corresponding column with texts, get the size
data_path = '/home/peng/Documents/NLP/Twitter_datasets/'

data_name = 'clean_url_twitter_29K.csv'
df = pd.read_csv(data_path + data_name, header = 0)

#get the size of text
num_docus = df['clean_url'].size

print df.columns.values
print num_docus
```

```python
df['clean_url'] = df['clean_url'].astype(str)
print df.dtypes
```

```python
#2. clean the data using pre
import Preprocessing_nlp as pre
clean_docus = []
for i in xrange(0, num_docus):
    
    if ((i+1)%1000 == 0):
        print "review %d of %d\n" % (i+1, num_docus)
    
    clean_docus.append(pre.review_to_words(df['clean_url'][i], filter_words = 'timeline'))
```

```python
#3. create bag of words....
from sklearn.feature_extraction.text import CountVectorizer

vectorizer = CountVectorizer(analyzer = 'word', \
                            tokenizer = None, \
                            preprocessor = None, \
                            stop_words = None, \
                            max_features = 10000)

train_data_features = vectorizer.fit_transform(clean_docus) #input should be lists

train_data_features = train_data_features.toarray()

print train_data_features.shape
```
