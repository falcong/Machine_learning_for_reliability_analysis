```python
#1.  import the raw data, note whether rul is included, if not 

import pandas as pd
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
%matplotlib inline

save_path = '/home/peng/Documents/Project_C/Turbofan/Dataset/'
names = xrange(0,26)
#df = pd.read_csv(save_path + 'train_FD001_rul.txt',header=0)
df = pd.read_csv(save_path + 'train_onlyone.txt',delim_whitespace=True, names=names, header=None)

#print df
#print df.describe()

# create a nan column to store the labels
df['target_easy']=np.nan
#df['ratio']=np.nan
print df
```

```python
## create label for ratio, the ratio of the length

for i in xrange(1, df[0].values.max()+1):
    
    #ratio_list.append(df[df[0]==i][1]/ len(df[df[0]==i]))
    #df[df[0]==i]['ratio']=(df[df[0]==i][1]/ len(df[df[0]==i]))
    df.loc[((df[0] == i) ), 'ratio']= df[df[0]==i][1]/ len(df[df[0]==i])
    #print i

```

```python
# label the top 30 and bottom 30 for the class healthy and failed 

for i in xrange(1, df[0].values.max()+1):
    df['target_easy'][df[df[0]==i].head(50).index.values] = 0
    df['target_easy'][df[df[0]==i].tail(50).index.values] = 1    
    
```

```python
#label two disc classes ratio


for i in xrange(1, df[0].values.max()+1):
    length = len(df[df[0]==i])
        
    #df[df['A'] > 2]['B'] = new_val  # new_val not set in df

    #df.loc[df[0] == i, 'target_easy'] []= new_val
    
    df.loc[((df[0] == i) & (df[1]<= int(0.2*length))), 'target_easy']=0
    
#    df.loc[((df[0] == i) & ( (df[1]>=int(0.4*length) ) & (df[1]<= int(0.6*length)))),
#           'target_easy']=1
    
    #df.loc[((df[0] == i) & ( (df[1]>=int(0.5*length) ) & (df[1]<= int(0.75*length)))),
    #       'target_easy']=2
    
    df.loc[((df[0] == i) & (df[1]>= int(0.8*length))), 'target_easy']=1
    
```

```python
#label three disc classes_ not ratio

for i in xrange(1, df[0].values.max()+1):

    
    length = len(df[df[0]==i])
    
    #df[df['A'] > 2]['B'] = new_val  # new_val not set in df

    #df.loc[df[0] == i, 'target_easy'] []= new_val
    
    df.loc[((df[0] == i) & (df[1]<= (length- 220))), 'target_easy']=0
    
    df.loc[((df[0] == i) & ( (df[1]>=(length - 200) ) & (df[1]<=(length - 150)))),
           'target_easy']=1
    
    df.loc[((df[0] == i) & ( (df[1]>=(length - 125) ) & (df[1]<= (length - 75)))),
           'target_easy']=2
    
    df.loc[((df[0] == i) & (df[1]>= (length - 50))), 'target_easy']=3
```

```python
#label three disc classes

for i in xrange(1, df[0].values.max()+1):
    length = len(df[df[0]==i])
    
    
    
    #df[df['A'] > 2]['B'] = new_val  # new_val not set in df

    #df.loc[df[0] == i, 'target_easy'] []= new_val
    
    df.loc[((df[0] == i) & (df[1]<= int(0.2*length))), 'target_easy']=0
    
    df.loc[((df[0] == i) & ( (df[1]>=int(0.4*length) ) & (df[1]<= int(0.6*length)))),
           'target_easy']=1
    
    #df.loc[((df[0] == i) & ( (df[1]>=int(0.5*length) ) & (df[1]<= int(0.75*length)))),
    #       'target_easy']=2
    
    df.loc[((df[0] == i) & (df[1]>= int(0.8*length))), 'target_easy']=2
```

```python
#label three classes

for i in xrange(1, df[0].values.max()+1):

    
    length = len(df[df[0]==i])
    
    #df[df['A'] > 2]['B'] = new_val  # new_val not set in df

    #df.loc[df[0] == i, 'target_easy'] []= new_val
    
    df.loc[((df[0] == i) & (df[1]<= int(0.333*length))), 'target_easy']=0
    
    df.loc[((df[0] == i) & ( (df[1]>=int(0.333*length) ) & (df[1]<= int(0.666*length)))),
           'target_easy']=1
    
    #df.loc[((df[0] == i) & ( (df[1]>=int(0.5*length) ) & (df[1]<= int(0.75*length)))),
    #       'target_easy']=2
    
    df.loc[((df[0] == i) & (df[1]>= int(0.666*length))), 'target_easy']=2
```

```python
# label four classes

for i in xrange(1, df[0].values.max()+1):

    
    length = len(df[df[0]==i])
    
    #df[df['A'] > 2]['B'] = new_val  # new_val not set in df

    #df.loc[df[0] == i, 'target_easy'] []= new_val
    
    df.loc[((df[0] == i) & (df[1]<= int(0.25*length))), 'target_easy']=0
    
    df.loc[((df[0] == i) & ( (df[1]>=int(0.25*length) ) & (df[1]<= int(0.5*length)))),
           'target_easy']=1
    
    df.loc[((df[0] == i) & ( (df[1]>=int(0.5*length) ) & (df[1]<= int(0.75*length)))),
           'target_easy']=2
    
    df.loc[((df[0] == i) & (df[1]>= int(0.75*length))), 'target_easy']=3
    
```

```python
# drop the useless columns 0 ,1 ,2, 27

df_drop = df.drop(df.columns[[0, 1]], axis=1)

# drop the nan and get the pure classes

df_1 = df_drop.dropna(axis = 0 , how='any')
#df_1 = df_1.sample(20000)
#print df_1
```
