```python
import pandas as pd
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
%matplotlib inline

save_path = '/home/peng/Documents/Project_C/Turbofan/'
names = xrange(0,26)

df = pd.read_csv(save_path + 'train_FD001.txt',names=names, header=None, delim_whitespace=True)

print df.columns.values.tolist()
```

```python
print len(df)
```

```python
rul_list = []
for i in xrange(0,len(df)):
    for j in xrange(1, df[0].values.max()+1):
        if df[0][i]==j:
            rul_list.append(len(df[df[0]==j]) - df[1][i])
    if i%100== 0:
        print "review %d of %d\n" % (i+1, len(df))
    
```

```python
print df[0].values.max()
```

```python
df['rul']=rul_list
```

```python
df.to_csv(save_path + 'train_FD001_rul.csv', header = True)
```
